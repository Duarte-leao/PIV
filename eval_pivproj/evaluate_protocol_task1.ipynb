```python
import matplotlib.pyplot as plt
from PIL import Image
import os

import scipy.io as sio
import numpy as np

def load_data(path, root, output_dir):
    data = sio.loadmat(os.path.join(output_dir, 'H_' + path.split('.')[0].split('rgb')[1] + '.mat'))
    return np.array(Image.open(os.path.join(root, path))), data['H']
    
# # SNS ----replace this entry -
# template_dir = '/DATASETS/formsns'

# input_dir = '/DATASETS/formsns/receitaSNS'
# output_dir = 'output'

# SNS ----replace this entry -
template_dir = 'D:/IST/MEEC/4ยบ Ano/P2/PIV/Projeto/Homography/template'

input_dir = 'D:/IST/MEEC/4ยบ Ano/P2/PIV/Projeto/Homography/input'
output_dir = 'D:/IST/MEEC/4ยบ Ano/P2/PIV/Projeto/Homography/output'

# Run project code
# os.system('python3 pivproject2022_task1.py %s %s %s' %(template_dir, input_dir, output_dir))
os.system('octave D:/IST/MEEC/4ยบ Ano/P2/PIV/Projeto/Homography/pivproject2022_task1.m %s %s %s' %(template_dir, input_dir, output_dir))
print('finished running code')
```

# Visualize the OUTPUT
This processing must be run with opencv installed and is only for visualization purposes

```python
import cv2
import glob

jpg_template=os.path.join(template_dir,glob.glob(template_dir + '/*.jpg')[0])

tm = Image.open(jpg_template)

for fname in sorted(os.listdir(input_dir)):
    if fname.endswith('.jpg') or fname.endswith('.png'):
        im, H = load_data(fname, input_dir, output_dir)

        # Apply homography to input image
        warped_im = cv2.warpPerspective(im, H, tm.size)
    
        # Show results
        plt.subplot(131)
        plt.imshow(im)
        plt.axis('off')
        plt.subplot(132)
        plt.imshow(tm)
        plt.axis('off')
        plt.subplot(133)
        plt.imshow(warped_im)
        plt.axis('off')
        plt.show()
```
